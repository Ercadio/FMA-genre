```python
from pydub import AudioSegment
from pydub.exceptions import CouldntDecodeError
import os
```

```python
directoryTree = [i for i in os.walk("fma_small_mp3")]

suites = [
    {
        "data": directoryTree[i][2],
        "suitePath": directoryTree[0][1][i - 1]
    } for i in range(1, 157)
]
for suite in suites:
    if not os.path.exists(".\\fma_small_wav\\" + suite["suitePath"]):
        os.makedirs(".\\fma_small_wav\\" + suite["suitePath"])
    for song in suite["data"]:
        try:
            audio_segment = AudioSegment.from_mp3("fma_small_mp3\\" + suite["suitePath"] + "\\" + song)
            #Export to .wav MONO
            audio_segment.export(".\\fma_small_wav\\" + suite["suitePath"] + "\\" + song.replace("mp3", "wav"),
                                format="wav")
        except CouldntDecodeError:
            print("There was a problem trying to decode the file " + song + ". Perhaps it is corrupted. Continuing...")
    print("Suite " + suite["suitePath"] + " completed...")
```
