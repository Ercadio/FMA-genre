```python
import matplotlib.pyplot as plt
from scipy.io import wavfile as wav
from scipy.fftpack import fft
import numpy as np
import winsound
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML, Image
```

```python
waveFile = "./fma_small_wav/000/000148.wav"
rate, data = wav.read(waveFile)
winsound.PlaySound(waveFile, winsound.SND_FILENAME)
```

```python
rc('animation', html='html5')
fig, ax = plt.subplots()

line1, line2 = ax.plot([], [], [], [], lw=1)
plt.close()

def init():
    line1.set_data([], [])
    line2.set_data([], [])
    ax.set_xlim(( 2000, 2100))
    ax.set_ylim((-5000, 5000))
    return (line1, line2)

def animate(i):
    view = data[i + 2000: i + 2100]
    x = range(i, 100 + i)
    y1 = [y0[0] for y0 in view]
    y2 = [y0[1] + 500 for y0 in view]
    line1.set_data(x, y1)
    line2.set_data(x, y2)
    ax.set_xlim((i, 100 + i))
    return (line1, line2)

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=True)
anim
```
